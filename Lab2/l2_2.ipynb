Steps for Handling Categorical Data

1. Import Libraries
2. Load Data
3. Seprate Input and Output attributes
4. Convert the categorical data into numerical data

```python
# Step 1: Import Libraries

import numpy as np 
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder

# Step 2: Load Data
        
datasets = pd.read_csv('https://raw.githubusercontent.com/Keval-Patel-1608/CE066_ML/main/Lab2/Exercise-CarData.csv') 
print("\nData :\n",datasets)
print("\nData statistics\n",datasets.describe())
```

```python
# Step 3: Seprate Input and Output attributes

# All rows, all columns except last 
X = datasets.iloc[:, 2:].values 
  
# Only last column  
Y = datasets.iloc[:, 1].values 

print("\n\nInput : \n", X) 
print("\n\nOutput: \n", Y) 
```

```python
le = LabelEncoder()
print("Before scaling ",X[ : ,4])
```

```python
X[ : ,4] = le.fit_transform(X[ : ,4])
FuelTypeData=X[ : ,4]
print("\n\nafter scaling of string data : \n", X[ :,4]) 
```

```python
# Step 4b: Use dummy variables from pandas library
#          to create one column for each country

dummy = pd.get_dummies(datasets['FuelType'])
print("\n\nDummy :\n",dummy)
datasets = datasets.drop(['FuelType','Weight'],axis=1)
datasets = pd.concat([dummy,datasets],axis=1)
print("\n\nFinal Data :\n",datasets)
```

```python
onehotencoder = OneHotEncoder()
#reshape the 1-D array to 2-D as fit_transform expects 2-D and finally fit the object 
x = onehotencoder.fit_transform(datasets.Price.values.reshape(-1,1)).toarray()
x.shape
```

```python
print(datasets.shape)
print(datasets.shape[1])
dfOneHot = pd.DataFrame(x, columns = ["HP_"+str(int(i)) for i in range(x.shape[1])]) 
df = pd.concat([datasets, dfOneHot], axis=1) #column
#droping the horsepower column 
df= df.drop(['HP'], axis=1) 
#printing to verify 
print(df.head())
```
